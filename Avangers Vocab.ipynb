```python
#Avengers Vocab !

import re
from collections import Counter


#    

def preprocess_text():
    # Avengers.txt  
    with open('Avengers.txt', 'r', encoding='utf-8') as f:
        text = f.read()

    #  
    text = text.lower()

    # 
    text = re.sub(r'[^a-z\s]', '', text)

    #   
    words = text.split()
    return words

def make_vocab():
    # 1.   
    words = preprocess_text()

    # 2.     
    word_count = {}
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1

    # 3.      +    
    def get_freq(item):
        return item[1]
    sorted_words = sorted(word_count.items(), key=get_freq, reverse=True) # 

    vocab = {}
    for idx, (word, _) in enumerate(sorted_words):     # 
        vocab[word] = idx

    return word_count, vocab



    #4.  input()   return  

def encoder():
    # vocab  
    _, vocab = make_vocab()

    #  
    text = input()

    #   
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    words = text.split()

    #   
    return [vocab.get(word, -1) for word in words]

if __name__ == "__main__":
    #   
    dic, vocab = make_vocab()
    print(dic)
    print(vocab)

    # encoder 
    result = encoder()
    print(result)




    #  :
    #       
    #         .
```
